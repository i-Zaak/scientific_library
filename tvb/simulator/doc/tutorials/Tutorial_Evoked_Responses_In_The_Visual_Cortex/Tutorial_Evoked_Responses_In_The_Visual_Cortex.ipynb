**Evoked Responses in the Visual Cortex**

```python
%pylab inlinefrom tvb.simulator.lab import *import numpyimport tvb.datatypes.projections as projectionsfrom scipy import io
```

==> 1. As usual, set up a Connectivity and its attributes.

```python
conn = connectivity.Connectivity()conn.speed = numpy.array([4.0])conn_coupling = coupling.Linear(a=0.042)
```

==> 2. Set up a model. In this configuration the topology of the phase portrait features a stable fixed point (a stable spiral) with a characteristics frequency of approximately 10Hz.You can check it using the nice phase_plane_interactive tool ...

```python
mod = models.Generic2dOscillator(a=-0.5, b=-15.0, c=0.0, d=0.02)
```

==> 3. Choose an integration scheme (noise or not?). Selecting a deterministic integration scheme allows you to observe very clearly the effect of stimulation-driven activity.With a stochastic scheme, you may observe some neural reponses which are not due to stimulation but to noise fluctuations (uncontrolled reponses of the resting state).

```python
hiss    = noise.Additive(nsig = numpy.array([0.015]))heunint = integrators.HeunStochastic(dt=2**-6, noise=hiss)
```

==> 4. Build a StimulusIn TVB's connectivity demo dataset region 35 and 36 represent areas V1 and V2 respectively in the left hemisphere. 

```python
conn.configure()nodes = [35, 36] stim_weights = numpy.zeros((conn.number_of_regions, 1))stim_weights[nodes] = numpy.array([3.5, 0.0])[:, numpy.newaxis]eqn_t = equations.PulseTrain()eqn_t.parameters["onset"] = 500.0 # mseqn_t.parameters["tau"]   = 5.0   # mseqn_t.parameters["T"]     = 500.  # 0.002kHz repetition frequencystimulus = patterns.StimuliRegion(temporal = eqn_t,                                  connectivity = conn,                                   weight = stim_weights)    
```

==> 5. Record the ouput as:

* Temporal Averaged time-series. By default only the first state variable is recorded for the `Generic2dOscillator` model.* EEG (FIXME. By default pr = projections.ProjectionRegionEEG() should do all the tricks)

```python
root_path = os.path.dirname(tvb.simulator.__file__)proj_mat_path = os.path.join(root_path,                                  'files', "connectivity",                                  "o52r00_irp2008", "projection_eeg_1020_62.mat")matlab_data = io.matlab.loadmat(proj_mat_path)eeg_projection = matlab_data["ProjectionMatrix"]    pr = projections.ProjectionRegionEEG()pr.projection_data = eeg_projection
```

```python
rec = (monitors.TemporalAverage(period=1e3 / 2048.), monitors.EEG(projection_matrix_data = pr, period=1e3/2048.)) 
```

==> 6. Create a `Simulator` instance. 

```python
def config_sim(stim=None):    sim = simulator.Simulator(model = mod,                           connectivity = conn,                          coupling = conn_coupling,                           integrator = heunint,                           monitors = rec,                          stimulus = stim)                        sim.configure()    return sim
```

==> 8. Create a `run` function to perform the simulation

```python
def run_simulation(sim):        LOG.info("Starting simulation...")    #Perform the simulation    tavg_data = []    tavg_time = []    eeg_data = []    eeg_time = []    for tavg, eeg in sim(simulation_length=2**10):    # approx 4 sec        if not tavg is None:            tavg_time.append(tavg[0])            tavg_data.append(tavg[1])                    if not eeg is None:            eeg_time.append(eeg[0])            eeg_data.append(eeg[1])                LOG.info("Finished simulation.")    return (tavg_data, tavg_time, eeg_data, eeg_time)
```

**Results**A. Activity in the absence of stimulation

```python
sim = config_sim()out = run_simulation(sim)
```

```python
TAVG = numpy.array(out[0])tt   = numpy.array(out[1])EEG  = numpy.array(out[2])
```

```python
subplot(211)plot(tt, TAVG[:, 0, nodes, 0])title("Temporal Averaged time-series")subplot(212)plot(tt, EEG[:, 0, 60, 0], 'k')title("EEG")
```

Figure 1: TAVG and EEG traces (0z) resting state

B. Stimulus-driven activity

```python
sim = config_sim(stim=stimulus)stout = run_simulation(sim)
```

```python
STAVG = numpy.array(stout[0])stt   = numpy.array(stout[1])SEEG  = numpy.array(stout[2])
```

```python
subplot(211)plot(stt, STAVG[:, 0, nodes, 0])title("Stim - Temporal Averaged time-series")subplot(212)plot(stt, SEEG[:, 0, 60, 0], 'k')title("Stim - EEG") 
```

Figure 2: TAVG and EEG traces (0Z) under stimulation -- onset is @ 500 ms
